Project: Spam Email Classifier

This project builds a simple machine learning model to classify text messages as spam or not spam. The model first converts raw text into numerical features using a bag-of-words approach, then trains a Multinomial Naive Bayes classifier to learn patterns associated with spam messages. After training, the model can predict whether a new unseen message is likely to be spam. This demonstrates the basic workflow of natural language processing (NLP) and supervised machine learning for text classification.

Libraries Used

pandas — for handling and organising the dataset

scikit-learn — for machine learning tools including:

train_test_split

CountVectorizer

MultinomialNB

accuracy_score

Accuracy Achieved

Initial toy dataset: 1.00 (100%)

Note: This perfect accuracy is likely due to the very small dataset size. Performance on larger real-world datasets is expected to be lower but more realistic.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
data = {
    "text": [
        "Win money now",
        "Claim your free prize",
        "Hello how are you",
        "Let's meet tomorrow",
        "Free entry in competition",
        "Are you coming to class",
        "Congratulations you won",
        "See you later"
    ],
    "label": [1, 1, 0, 0, 1, 0, 1, 0]  # 1 = spam, 0 = not spam
}

df = pd.DataFrame(data)
df
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["label"], test_size=0.25, random_state=42
)
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
model = MultinomialNB()
model.fit(X_train_vec, y_train)
predictions = model.predict(X_test_vec)
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
my_message = ["I want to steal your number"]
my_message_vec = vectorizer.transform(my_message)
prediction = model.predict(my_message_vec)

if prediction[0] == 1:
    print("Spam detected!")
else:
    print("Not spam.")
```
